```python
import folium
from folium import plugins
import pandas as pd
well_data=pd.read_csv('Wells2.csv')
print(well_data)
import os
print(os.getcwd())

#create base map
m = folium.Map(location=[19.7011, 76.8967], tiles="Stamen Terrain",zoom_start=12,width=800,height=400,control_scale=True)

# add tiles to map
folium.raster_layers.TileLayer('Open Street Map').add_to(m)
folium.raster_layers.TileLayer('Stamen Terrain').add_to(m)
folium.raster_layers.TileLayer('Stamen Toner').add_to(m)
folium.raster_layers.TileLayer('Stamen Watercolor').add_to(m)
folium.raster_layers.TileLayer('CartoDB Positron').add_to(m)
folium.raster_layers.TileLayer('CartoDB Dark_Matter').add_to(m)


# add geojson file to map
folium.GeoJson('well_boundary.geojson', name='Well_boundary ').add_to(m)


# add layer control to show different maps
folium.LayerControl().add_to(m)

# plugin for mini map
minimap = plugins.MiniMap(toggle_display=True)

# add minimap to map
m.add_child(minimap)

# add scroll zoom toggler to map
plugins.ScrollZoomToggler().add_to(m)

# add full screen button to map
plugins.Fullscreen(position='topright').add_to(m)


#creating objects from dataframe 
lat=list(well_data["Lat_DD"])
lon=list(well_data["Long_DD"])
Wellname=list(well_data["Well_code"])
WellDepth=list(well_data["Depth_Of_Well(m)"])
WellDepth1=str(WellDepth)
WellTopLevel=list(well_data["Altitude_of_WT(m)"])
WellTopLevel1=str(WellTopLevel)
plot=list(well_data["month_plot"])
picture=list(well_data["well_pic"])

#to know data type of different objects generated
print(type(Wellname))
print(type(WellDepth))
print(type(WellDepth1))
print(type(WellTopLevel))
print(type(WellTopLevel1))
print(type(plot))
print(type(picture))

#Adding Marker

for lt,ln,nm,dep,lev,pl,pic in zip(lat,lon,Wellname,WellDepth,WellTopLevel,plot,picture):

 folium.Marker(location=[lt,ln],popup="<b>Wellname  : </b>"+nm+ " <br> <b>WellDepth: </b> "+str(dep)+ 
               "<br> <b>WellTopLevel : </b> "+str(lev)+"<br> <img src="+pl+" height=200 width=350>"+"<br> <img src="+
               pic+" height=200 width=350>",tooltip='click for more info',icon=folium.Icon(color='green')).add_to(m)

# measure control
measure_control = plugins.MeasureControl(position='topleft', 
                                         active_color='red', 
                                         completed_color='red', 
                                         primary_length_unit='kilometers')

# add measure control to map
m.add_child(measure_control)

# draw tools
draw = plugins.Draw(export=True)

# add draw tools to map
draw.add_to(m)


# display map
m.add_child(folium.LatLngPopup())
m
#save map
#m.save("map2.html")

```
